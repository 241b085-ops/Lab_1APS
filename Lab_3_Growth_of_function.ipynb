# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)
def linear_time_difference(numbers):
  quotient = 0
  for values in numbers:
    quotient -= values
  return quotient

n_values = list(range(1,100))
time_values =[]

for n in n_values:
  data = list(range(n))
  linear_time_difference(data)
  time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
def logarithmic_time_sum(numbers):
   total = 0
   for value in numbers:
        total += value
   return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Logarithmic Time Complexity")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
def log_linear_time_sum(numbers):
  total = 0
  for values in numbers:
    total += values
  return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
  data = list(range(n))
  linear_time_sum(data)
  time_values.append(n)

plt.figure()
plt.plot(n_values,time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Log_Linear Time Complexity")
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time. plot the graph as well, similar to above example

def polynomial_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = list(range(1, 51))
time_values = []

for n in n_values:
    time_values.append(n*n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (operations)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
def exponential_time_example(n):
  count = 0
  for i in range(n):
    for j in range(2**i):
      count +=1
      return count
n_values = list(range(1,51))
time_values = []
for n in n_values:
  time_values.append(2**n)
plt.figure()
plt.plot(n_values,time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (operations)")
plt.title("O(2^n) Exponential Time Complexity")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
n_values = np.arange(1, 31)
y_constant = np.ones_like(n_values)
y_log = np.log2(n_values)
y_linear = n_values
y_n_log_n = n_values * np.log2(n_values)
y_n_squared = n_values**2
y_two_power_n = 2**n_values

plt.figure(figsize=(12, 8))

plt.plot(n_values, y_constant, label='O(1) Constant')
plt.plot(n_values, y_log, label='O(log n) Logarithmic')
plt.plot(n_values, y_linear, label='O(n) Linear')
plt.plot(n_values, y_n_log_n, label='O(n log n) Log-Linear')
plt.plot(n_values, y_n_squared, label='O(n^2) Polynomial')

plt.plot(n_values, y_two_power_n, label='O(2^n) Exponential')

plt.xlabel("Input size (n)")
plt.ylabel("Operations / Time")
plt.title("Comparison of Time Complexities")

plt.yscale('log')
plt.grid(True, which="both", ls="-", axis='y', alpha=0.5)
plt.legend()
plt.ylim(1, 10**10)
plt.xlim(1, 30)
plt.show()
```

```python
import time
import random
import math
import matplotlib.pyplot as plt
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def measure_time_complexity(func, n_values):
    times = []
    for n in n_values:

        data = [random.randint(0, n) for _ in range(n)]

        start_time = time.perf_counter_ns()
        func(data)
        end_time = time.perf_counter_ns()

        times.append((end_time - start_time) / 1_000_000)
    return times
n_values_qs = [10, 50, 100, 200, 500, 1000, 2000, 5000, 10000]

actual_times_qs = measure_time_complexity(quick_sort, n_values_qs)
theoretical_log_n = [math.log2(n) for n in n_values_qs]
theoretical_n_log_n = [n * math.log2(n) for n in n_values_qs]


if theoretical_n_log_n[-1] > 0:
    scaling_factor_n_log_n = actual_times_qs[-1] / theoretical_n_log_n[-1]
else:
    scaling_factor_n_log_n = 1.0

if theoretical_log_n[-1] > 0:
    scaling_factor_log_n = actual_times_qs[-1] / theoretical_log_n[-1] / n_values_qs[-1]
else:
    scaling_factor_log_n = 1.0

theoretical_log_n_scaled = [t * scaling_factor_log_n for t in theoretical_log_n]
theoretical_n_log_n_scaled = [t * scaling_factor_n_log_n for t in theoretical_n_log_n]
plt.figure(figsize=(12, 8))
plt.plot(n_values_qs, actual_times_qs, marker='o', linestyle='-', label='Quick Sort (Actual Time)')
plt.plot(n_values_qs, theoretical_n_log_n_scaled, linestyle='--', color='red', label='O(n log n) Theoretical (Scaled)')
plt.plot(n_values_qs, theoretical_log_n_scaled, linestyle=':', color='green', label='O(log n) Theoretical (Scaled)')

plt.xlabel("Input size (n)")
plt.ylabel("Time (milliseconds)")
plt.title("Quick Sort Time Complexity Comparison")
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

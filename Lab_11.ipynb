# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import itertools

def tsp_dynamic_programming(graph, start=0):
    n = len(graph)
    DP = {}


    for k in range(n):
        if k == start:
            continue
        DP[(frozenset([start, k]), k)] = (graph[start][k], start)


    for subset_size in range(3, n + 1):

        for subset in itertools.combinations(range(n), subset_size):
            if start not in subset:
                continue
            subset_frozen = frozenset(subset)


            for j in subset:
                if j == start:
                    continue

                min_cost, min_prev = float('inf'), None


                for k in subset:
                    if k == j or k == start:
                        continue

                    prev_subset = subset_frozen - {j}

                    cost_k, _ = DP[(prev_subset, k)]


                    cost = cost_k + graph[k][j]

                    if cost < min_cost:
                        min_cost, min_prev = cost, k

                DP[(subset_frozen, j)] = (min_cost, min_prev)


    min_total_cost, min_last_city = float('inf'), None
    full_set = frozenset(range(n))
    for j in range(n):
        if j == start:
            continue
        cost_j, _ = DP[(full_set, j)]
        cost = cost_j + graph[j][start]
        if cost < min_total_cost:
            min_total_cost, min_last_city = cost, j


    path = []
    subset_for_path = full_set
    current_city = min_last_city

    while current_city != start:
        path.append(current_city)
        _, prev_city = DP[(subset_for_path, current_city)]
        subset_for_path = subset_for_path - {current_city}
        current_city = prev_city

    path.append(start)
    path.reverse()
    path.append(start)

    return min_total_cost, path

n = int(input("Enter number of cities: "))
graph = []
print("Enter adjacency matrix row by row:")
for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)


cost, path = tsp_dynamic_programming(graph, start=0)

print("Minimum travelling cost:", cost)
print("Path of cities visited:", " → ".join(str(p + 1) for p in path))
```

# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    all_nodes = set(graph.keys())
    for node_key in graph:
        for neighbor_node in graph[node_key]:
            all_nodes.add(neighbor_node)

    distances = {}
    visited = []
    for node in all_nodes:
        distances[node] = float('inf')

    if start not in all_nodes:
        return {node: float('inf') for node in all_nodes}
    distances[start] = 0

    while len(visited) < len(all_nodes):

        min_node = None
        min_distance = float('inf')
        for node in all_nodes:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:

            break

        visited.append(min_node)
        if min_node in graph:
            for neighbor, weight in graph[min_node].items():
                if neighbor not in visited: # Ensure neighbor is not already visited
                    new_distance = distances[min_node] + weight
                    if new_distance < distances[neighbor]:
                        distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
graph = {}
n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input(f"Enter name for node {i+1}: ")

    if node not in graph:
        graph[node] = {}

    while True:
        try:
            edges = int(input(f"Enter the NUMBER of neighbors for node '{node}': "))
            break
        except ValueError:
            print("Invalid input. Please enter a single integer for the number of neighbors.")

    for j in range(edges):
        neighbor = input(f"Enter neighbor {j+1} for '{node}': ")
        while True:
            try:
                weight_input = input(f"Enter weight for edge from '{node}' to '{neighbor}': ")
                weight = int(weight_input)
                break
            except ValueError:
                print("Invalid input. Please enter a single integer for weight.")

        graph[node][neighbor] = weight

        if neighbor not in graph:
            graph[neighbor] = {}
        graph[neighbor][node] = weight
```

```python
def dijkstra_with_path(graph, start):

    distances = {}
    visited = []
    previous = {}


    for node in graph:
        distances[node] = float('inf')
        previous[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')


        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)


        for neighbor, weight in graph[min_node].items():
            if neighbor not in visited:
                new_distance = distances[min_node] + weight

                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance
                    previous[neighbor] = min_node   # Track path

    return distances, previous
def get_path(previous, start, end):
    path = []
    current = end

    while current is not None:
        path.append(current)
        current = previous[current]

    path.reverse()
    return path
```

```python

print("\n--- Shortest paths from A ---")
distances_A, previous_A = dijkstra_with_path(graph, 'A')
for node in distances_A:
    print(f"Distance from A to {node} = {distances_A[node]}")
    if distances_A[node] != float('inf'):
        path = get_path(previous_A, 'A', node)
        print(f"Path: {' -> '.join(path)}")
print("\n--- Shortest paths from B ---")
distances_B, previous_B = dijkstra_with_path(graph, 'B')
for node in distances_B:
    print(f"Distance from B to {node} = {distances_B[node]}")
    if distances_B[node] != float('inf'):
        path = get_path(previous_B, 'B', node)
        print(f"Path: {' -> '.join(path)}")
print("\n--- Shortest paths from F ---")
distances_F, previous_F = dijkstra_with_path(graph, 'F')
for node in distances_F:
    print(f"Distance from F to {node} = {distances_F[node]}")
    if distances_F[node] != float('inf'):
        path = get_path(previous_F, 'F', node)
        print(f"Path: {' -> '.join(path)}")
```
